```python
import requests
import os
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv('API_KEY')

query = "grmgtijyti"
url = "https://marvelrivalsapi.com/api/v2/player/{query}/match-history"
headers = {"x-api-key": api_key}

response = requests.get(url, headers=headers, params={"season": 1.5})
print(response.status_code)
print(response.json())
```
